Imbalanced Dataset:

A Dataset with an unequal class distribution

```python
import pandas as pd
import numpy as np
```

```python
credit = pd.read_csv('../../Datasets/credit_data.csv')
```

```python
credit.head()
```

```python
credit['Class'].value_counts()
```

This is Highly Imbalanced DataSet (Too much difference in numbers)

0 ---> Legal Transaction

1 ---> Illegal Transaction

```python
#separating the legit and fraudlent trransactions
legit = credit[credit.Class == 0]
fraud = credit[credit.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

Under-Sampling

Build a sample dataset containing similar distribution of legit and fraudlent transaction

no of fraudlent transactions --> 492

```python
legit_sample = legit.sample(n=492)
```

```python
print(legit_sample.shape)
```

Now Both are same number

So we concatenate the two dataframes

```python
new_dataset = pd.concat([legit_sample, fraud], axis = 0)
```
