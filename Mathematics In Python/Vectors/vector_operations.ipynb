# Plotting a vector


```python
plt.quiver(0,0,4,5)
plt.show()
```

```python
plt.quiver(0,0,4,5, scale_units = 'xy', angles='xy', scale=1)
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show
```

```python
plt.quiver(0,0,4,5, scale_units = 'xy', angles='xy', scale=1, color = 'b')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show
```

# Addition of Two Vectors

```python
vector_1 = np.asarray([0,0,2,3])

vector_2 = np.asarray([0,0,3,-2])

sum = vector_1 + vector_2

print(sum)
```

```python
plt.quiver(0,0,2,3, scale_units = 'xy', angles='xy', scale=1)
plt.quiver(0,0,3,-2, scale_units = 'xy', angles='xy', scale=1, color='r')
plt.quiver(0,0,5,1, scale_units = 'xy', angles='xy', scale=1, color='y')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show
```

# Subtraction of two vectors

```python
vector_1 = np.asarray([0,0,2,3])

vector_2 = np.asarray([0,0,3,-2])

difference = vector_1 - vector_2

print(difference)
```

```python
plt.quiver(0,0,2,3, scale_units = 'xy', angles='xy', scale=1)
plt.quiver(0,0,3,-2, scale_units = 'xy', angles='xy', scale=1, color='r')
plt.quiver(0,0,-1,5, scale_units = 'xy', angles='xy', scale=1, color='y')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show
```

# Multiplying a vector by a scalar

```python
vector_1 = np.asarray([0,0,2,3])

multiply = 2*vector_1

print(multiply)
```

```python
plt.quiver(0,0,2,3, scale_units='xy', angles='xy',scale=1,color='r')
plt.quiver(0,0,4,6, scale_units='xy', angles='xy',scale=1,color='g')
plt.xlim(-8,8)
plt.ylim(-8,8)
```

```python
plt.quiver(0,0,2,3, scale_units='xy', angles='xy',scale=1,color='r')
plt.quiver(0,0,-1,-1.5, scale_units='xy', angles='xy',scale=1,color='g')
plt.xlim(-8,8)
plt.ylim(-8,8)
```

# Dot Product of a vector

```python
plt.quiver(0, 0, 2, 3, scale_units="xy", angles="xy", scale=1, color='r')
plt.quiver(0, 0, 2,-3, scale_units="xy", angles="xy", scale=1, color='b')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
vector_1 = np.asarray([0,0,2,3])
vector_2 = np.asarray([0,0,4,4])

dot_prod = np.dot(vector_1,vector_2)

print(dot_prod)
```

# Cross product of two vectors

```python
vector_1 = np.asarray([2,3])
vector_2 = np.asarray([4,4])

cross_prod = np.cross(vector_1,vector_2)

print(cross_prod)
```

```python
plt.quiver(0,0,2,3, scale_units='xy', angles='xy',scale=1,color='r')
plt.quiver(0,0,4,4, scale_units='xy', angles='xy',scale=1,color='g')
plt.quiver(0,0,0,-4, scale_units='xy', angles='xy',scale=1,color='y')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
vector_3 = np.asarray([5,10,20])

vector_4 = np.asarray([18,32,50])

cross_product = np.cross(vector_3,vector_4)

print(cross_product)
```

# Projection of 'a' vector on 'v' vector

```python
a = np.asarray([2, 5])
v = np.asarray([8, -6])

# magnitude of vector "v"
mag_v = np.sqrt(np.sum(v**2))

proj_a_on_v = (np.dot(a, v) / mag_v**2) * v

print(proj_a_on_v)
```
